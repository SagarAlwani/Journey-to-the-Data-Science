How can i filter rows of a pandas data frame using column values?

```python
import pandas as pd
```

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

```python
movies.shape
```

```python
list_booleans = []
for length in movies.duration:
    if length >= 200:
        list_booleans.append(True)
    else:
        list_booleans.append(False)
```

```python
list_booleans[0:5]
```

```python
len(list_booleans)
```

```python
#Convert booleans list into pandas series
is_long = pd.Series(list_booleans)
```

```python
#Remember these two are different:
movies['genre']
```

```python
#and this
movies[is_long]
```

```python
type(movies['genre'])
```

```python
type(movies[is_long])
```

```python
#Instead of all above we can do simple is
is_long = movies.duration >= 200
is_long.head()
```

```python
movies[is_long]
```

```python
#It also worked
movies[movies.duration >= 200]
```

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.loc[(movies.duration >= 200) & (movies.genre == 'Crime'),'genre']
```

```python
movies.loc[(movies.duration >= 200) & (movies.genre == 'Crime')]
```
